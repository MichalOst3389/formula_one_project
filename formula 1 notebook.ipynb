```python
import pandas as pd #importing pandas
import seaborn as sns# importing seaborn
import matplotlib as plt# importing matplotlib
```

```python
allrace_df = pd.read_csv('AllRace.csv')#import AllRace.csv
constructor_standings_df = pd.read_csv('ConstructorStandings.csv')#import cunstructorStandings.csv
driver_standings_df = pd.read_csv('DriversStandings.csv')#import DriverStandings.csv
```

```python
driver_standings_df
```

```python
#which driver has the most overall points
```

```python
driver_points_df = driver_standings_df.groupby('Driver')['PTS'].sum().reset_index()# groupby drivers, sum of total points
```

```python
driver_points_df = driver_points_df.sort_values(by = 'PTS', ascending = False)
```

```python
driver_points_df.head(25)#Sum of driver points for top 25 drivers
```

```python
sns.barplot(x = 'PTS', y = 'Driver', data = driver_points_df.head(25))
```

```python
driver_points_per_race_avg = driver_standings_df.groupby('Driver')['PTS'].mean().reset_index()
```
